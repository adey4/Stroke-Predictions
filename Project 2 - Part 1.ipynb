```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = 'healthcare-dataset-stroke-data.csv'
df = pd.read_csv(data, index_col='id')
df.head()
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df.info()
```

```python
corr = df.corr()

sns.heatmap(corr, cmap='coolwarm', annot=True)
plt.show()
```

1. Data is from [Kaggle](https://www.kaggle.com/datasets/fedesoriano/stroke-prediction-dataset)
2. The dataset contains 11 clinical features for predicting strokes in patients. It contains 5110 data points with 11 attributes, including the target variable 'stroke.'
3. 'stroke' is the binary target variable representing whether the patient had a stroke (stroke = 1) or not (stroke = 0).
4. This is a binary classification problem.
5. There are 11 features.
6. There are 5110 rows.
7. Challenges include handling missing data in the 'bmi' column and model choice to account for low correlation between features and target variable.
